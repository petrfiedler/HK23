```python
pd.set_option('display.max_rows', 45)
df = pd.DataFrame(dc_file.dtypes, columns=['dtype'])
df['null'] = dc_file.isnull().sum()
df['null_pct'] = dc_file.isnull().sum() / dc_file.shape[0]
df['null_pct'] = df['null_pct'].apply(lambda x: '{:.1%}'.format(x))
df['nunique'] = dc_file.nunique()
df['unique'] = dc_file.apply(lambda x: x.unique() if x.nunique() < 5 else "[...]")
df
```
