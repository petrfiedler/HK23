```python
dc_file['timestamp'] = dc_file['timestamp'].astype('datetime64[ns]')
```

```python
dc_file_mean = dc_file.groupby(pd.Grouper(key='timestamp', freq='1min')).mean()
dc_file_sum = dc_file.groupby(pd.Grouper(key='timestamp', freq='1min')).sum()
dc_file_mean.dropna(inplace=True)
dc_file_sum = dc_file_sum.loc[dc_file_mean.index]
```

```python
dc_file_mean.reset_index(inplace=True)
dc_file_sum.reset_index(inplace=True)
```
